# 공연/행사 분야 페이지 정보 가져오기


```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import requests as req
from bs4 import BeautifulSoup as bs 
import time
import pandas as pd
import numpy as np
```

### 공연/행사 페이지 내에서 요소 추출 함수

```python
def content_extract_event():
    page_now = driver.page_source
    soup = bs(page_now, "lxml")
    # 파일 변환 후 행사 명, 설명, 기간 가져오기 
    title = soup.select("#topTitle")[0].text 
    info = soup.select("#topCp > div > h3 > em")[0].text
    period_start = soup.select("div.inr_wrap > div > ul > li:nth-child(1) > span")[0].text
    period_end = soup.select("div.inr_wrap > div > ul > li:nth-child(2) > span")[0].text
    location = soup.select("div.inr_wrap > div > ul > li:nth-child(6) > span")[0].text
    page_url = driver.current_url
    # 글 하나의 contents를 딕셔너리로 변환 후 리스트 저장.
    performance = {"title": title, 
                "contents": info,
                "period": (period_start + " ~ " + period_end),
                "location":location,
                "page_url":page_url
               }    
    performance_list.append(performance)
    # 저장 후 뒤로 가기 
    
    driver.back()    
    time.sleep(1)
```

# 최종 가져오기

```python
# 최종 담길 리스트
performance_list = []
```

```python
performance_url = "https://korean.visitkorea.or.kr/list/travelinfo.do?service=show" #공연/행사 url
driver = wb.Chrome()
driver.get(performance_url)
# 일단 1-5 만 하는 걸로

for i in range(2, 11): #페이지 총 넘김 수 
    # 첫 번째 글만 li.bdr_nor 고정 
    time.sleep(1.5)
    li_content = driver.find_element(By.CSS_SELECTOR, "#contents > div.wrap_contView.clfix > div.box_leftType1 > ul > li.bdr_nor > div.area_txt > div > a")
    li_content.send_keys(Keys.ENTER)
    # 상세 페이지 들어가기
    time.sleep(1.5)
    # 함수 돌리기 
    content_extract_event()
    # 두 번째 글 부터, li 
    for j in range(2, 11):
        time.sleep(1.5)
        li_content = driver.find_element(By.CSS_SELECTOR, f"#contents > div.wrap_contView.clfix > div.box_leftType1 > ul > li:nth-child({j}) > div.area_txt > div > a")
        li_content.send_keys(Keys.ENTER)
        # 상세 페이지 들어가기
        time.sleep(2)
        content_extract_event()
   # 한 페이지 넘기기 - 6~10페이지
    if i == 6: 
        next = driver.find_element(By.CSS_SELECTOR, "#contents > div.wrap_contView.clfix > div.box_leftType1 > div.page_box>a:nth-child(6)")
        next.send_keys(Keys.ENTER)
        time.sleep(1.5)
        
    if i >= 6: 
        next = driver.find_element(By.CSS_SELECTOR, f"#contents > div.wrap_contView.clfix > div.box_leftType1 > div.page_box>a:nth-child({i-3})" )
        next.send_keys(Keys.ENTER)
        time.sleep(1.5)
    # 한 페이지 넘기기 - 1~5페이지 
    elif i <= 5:
        next = driver.find_element(By.CSS_SELECTOR, f"#contents > div.wrap_contView.clfix > div.box_leftType1 > div.page_box>a:nth-child({i})" )
        next.send_keys(Keys.ENTER)
        time.sleep(1.5) 
```

# csv파일로 변환하기 

```python
performance_file = pd.DataFrame.from_dict(performance_list)
performance_file.index = performance_file.index+1
performance_file.to_csv("performance_list.csv", encoding="utf-8-sig")
```
