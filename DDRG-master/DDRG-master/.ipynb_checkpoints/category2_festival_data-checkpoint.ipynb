# 축제 데이터 가져오기

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import requests as req
from bs4 import BeautifulSoup as bs 
import time
import pandas as pd
import numpy as np
```

### 상세페이지 내용 긁어오기 함수 만들기

```python
festival_list = []
```

```python
# 반복하는 부분 함수 만들기 
def content_extract_festival():
    page_now = driver.page_source
    soup = bs(page_now, "lxml")
    # 파일 변환 후 행사 명, 설명, 기간 등 가져오기 
    title = soup.select("#festival_head")[0].text 
    try:
        info = soup.select("section.poster_title > div > div > div.festival_title > span")[0].text
    except:
        info = ""
    period = soup.select("div.poster_detail_wrap > div > div.img_info_box > ul > li:nth-child(1) > p")[0].text
    location = soup.select("div.poster_detail_wrap > div > div.img_info_box > ul > li:nth-child(2) > p")[0].text
    page_url = driver.current_url
    # 페스티벌 하나의 contents를 딕셔너리로 변환 후 리스트 저장.
    festival = {"title": title, 
                "contents":info,
                "period": period,
                "location":location,
                "page_url":page_url
               }  
    festival_list.append(festival)

    # 저장 후 뒤로 가기 
    driver.back()
    time.sleep(2)
```

# 개최중

```python
festival_url = "https://korean.visitkorea.or.kr/kfes/list/wntyFstvlList.do" #축제 url 
driver = wb.Chrome()
driver.get(festival_url)

# 개최중만 검색 
classify_period = driver.find_element(By.CSS_SELECTOR, "#searchDate > option:nth-child(2)")
classify_period.click()
search_btn = driver.find_element(By.CSS_SELECTOR, "#btnSearch")
search_btn.send_keys(Keys.ENTER)
time.sleep(1)
# 개최중인 축제 리스트 볼 수 있게 END키 사용
for i in range(9):
    body = driver.find_element(By.CSS_SELECTOR, "body")
    body.send_keys(Keys.END)
    time.sleep(1)
# 축제 갯수는 최대 100개
for j in range(1, 101):
    try:
        li_content = driver.find_element(By.CSS_SELECTOR, f"#fstvlList > li:nth-child({j}) > a")
        li_content.click()
        content_extract_festival()
    except:
        break
```
